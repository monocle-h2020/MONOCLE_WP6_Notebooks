## Introduction
This demonstration shows how in situ hyperspectral reflectance observations fitted using atmospheric models can be used to determine drift over time in the downwelling irradiance response. 

We will use data collected using an Rflex system, a precursor of the MONOCLE So-Rad solar-tracking hyperspectral radiometry platform, which keeps the sensors close to an optimal 135 degrees azimuth relative to the solar azimuth. The data were collected at the Finnish Environment Institute (SYKE) and we will be looking at data collected on the _Finnlines_ merchant vessel _Finnmaid_ during 2015. This is a suitable data set for our purposes, because the sensors were exposed for many months at a time, regularly cleaned and calibrated annually. 

While the sensors were at sea, weather and illumination conditions will not always have been appropriate to measure water-leaving reflectance. Fortunately several algorithms now exist to automatically fit Rrs spectra from the input (ir)radiance, allowing us to remove suboptimal observations based on fitting uncertainty. Here we use the 3C model of Gege et al. 2016 as implemented and validated by Groetsch et al. (2017).[References](#References) are given below.

The sensors were clean and lab-calibrated prior to the first day of deployment. So it is possible to look for signs of degradation over time  in the Ed sensor records compared to expected values. Signal degradation can be sudden (extreme biofouling) or gradual (biofouling, sensor degradation), so we need to observe sensor records over a period of time to determine any gradual drift.

We adopted the 3C model fitting routine from https://gitlab.com/pgroetsch/rrs_model_3C/-/blob/master/rrs_model_3C.py


```python
# Import dependencies
import os
import pandas as pd
import matplotlib.pyplot as pl
import datetime
import numpy as np
from scipy import stats
import rrs_signal_trends
import rrs_signal_trends.rrs_demo_backend as backend
import rrs_signal_trends.edfit_functions as edfit
print("Dependencies imported")

# For inline plots
%matplotlib notebook 
```

## Modelled and observed components

Below we use a small subset of the data to introduce the models we are using. Let's plot some of the modelled and observed data.


```python
#### read sample data
data = pd.read_csv('./rrs_signal_trends/shipdata.csv', index_col=0)

r = data.iloc[9,:]  # there are 10 samples included in this demontration, pick any from 0-9
result = edfit.process_single(r)  # this calls the data fitting routine, not discussed here
wavelength = edfit.wlin

##### plot results
fig = pl.figure()
p1 = pl.subplot(121)
p1.plot(wavelength, result['Lt']/result['Ed'], label='Lu/Ed obs. [sr$^{-1}$]')
p1.plot(wavelength, result['Lu_Ed_mod'], label='Lu/Ed modelled')
p1.plot(wavelength, result['Rrs_obs'], label='R$_{rs}$ obs. [sr$^{-1}$]')
p1.plot(wavelength, result['Rrs_mod'], label='R$_{rs}$ modelled [sr$^{-1}$]')

p2 = pl.subplot(122)
p2.plot(wavelength, result['Ed'], label='E$_d$ obs.')
p2.plot(wavelength, result['Ed_GC90'], label='E$_d$ GC90')
p1.set_xlabel('Wavelength, [nm]')
p1.set_ylabel('Remote Sensing Reflectance, [sr-1]')
p2.set_ylabel('Ed [W/m2 nm]')
p1.set_xlim(wavelength[0], wavelength[-1])
p2.set_xlim(wavelength[0], wavelength[-1])
p1.legend(loc=0)
p2.legend(loc=0)
pl.tight_layout()
```

The left plot shows that the main reflectance components (sky reflectance (Lu/Ed) and water-leaving reflectance (Rrs)) were  accurately approximated by the 3C model (which is not the case for all samples in the dataset). This means that the 3C model has managed to fit an atmospheric model to the data. As a measure of the goodness-of-fit, we will use the slope of the linear regression between the modelled and the observed reflectance. 

The right plot uses some of the 3C model optimised model coefficients to compute the full spectrum of downwelling irradiance (Ed) using the model by Gregg and Carder (1990) abbreviated here as GC90. The observed Ed and modelled Ed do not completely correspond due to uncertainties in the model parameterisation, notably the relative humidity, air pressure and temperature. These were not recorded by the deployed system, but could be quite useful additions! 

## Data filtering

Illumination conditions change quite frequently during most days, due to changing cloud cover. Sensor sensitivity and fouling, on the other hand, are expected to change gradually (or _very_ abrubptly in case of heavy fouling). Our first task is to make an appriate data selection to allow inspection of trends in signal quality over time. We want to be able to predict the expected Ed as well as possible, so this comes down to finding the sunny moments in the data set. 

First we will have a look at all daytime results to determine appropriate filter thresholds. Note that modelled results were saved to a database because fitting a year's worth of data does take a while!

```python
## Select all daytime results 
sunzenithmax = 80.0  # maximum allowed sun zenith angle (quality control, ensure sufficient light intensity)
c3fitmin = 0.9       # minimum selection threshold for fitting of Rrs to observation (quality control)
Ed400min = 20        # minimum downwelling irradiance intensity (ensure sufficient light intensity)
lsed400max = 1.5     # maximum value of the Lsky/Ed ratio [1/sr] at 400 nm, to select clear-sky conditions

results = backend.query_db(backend.dbpath, sunzenithmax=sunzenithmax,
                   c3fitmin=c3fitmin, Ed400min=Ed400min, lsed400max=lsed400max)

# read timestamps, day-of-year, and days since deployment for plotting purposes
times = [datetime.datetime.strptime(r['timestamp'], '%Y-%m-%d %H:%M:%S') for r in results]
doy = [t.timetuple().tm_yday for t in times]
deployment_doy = np.min(doy)
dep_day = [d-deployment_doy for d in doy]
Ed400 = [r['Ed400'] for r in results]  # downwelling irradiance at 400 nm
LsEd400 = [r['Ls_Ed_400'] for r in results]  # ratio of sky radiance over downwelling irradiance ratio at 400nm

# plot results
pl.figure()
pl.subplot(2,1,1)
pl.plot(times, Ed400, '.b', markersize=2, alpha=0.5, label='Ed(400) (brightness)')  # irradiance over time
pl.ylabel('Ed(400) [W/m2 nm]')
_ = pl.legend(loc=1)
pl.subplot(2,1,2)
pl.plot(times, LsEd400, '.g', markersize=2, alpha=0.5, label='Ls/Ed(400) (cloudiness)')  # cloudiness over time
pl.ylabel('Ls(400)/Ed(400) [1/sr]')
pl.xlabel('Time')
_ = pl.legend(loc=1)
```

As expected, we can see high variability in the illumination conditions during the day (try zooming in on a single day). There is also seasonality visible in the highest intensity of downwelling irradiance reached. 

To make useful comparisons between modelled and observed light intensities measured by the sensors we need to select only results corresponding to blue skies and with the sun overhead because other conditions cannot be verified using the model. Let's try to find suitable threshold values for solar elevation, downwelling irradiance intensity and cloudiness. For the latter we look at the Lsky/Ed ratio, with lower values indicating clearer skies. 

In the code below you can vary these selection criteria to arrive at a dataset representing the most sunny days.


```python
# select data with appropriate thresholds for sunny conditions
sunzenithmax = 50.0  # maximum allowed sun zenith angle (quality control, ensure sufficient light intensity)
c3fitmin = 0.9       # minimum selection threshold for fitting of Rrs to observation (quality control)
Ed400min = 500       # minimum downwelling irradiance intensity (ensure sufficient light intensity)
lsed400max = 0.006   # maximum value of the Lsky/Ed ratio [1/sr] at 400 nm, to select clear-sky conditions
results = backend.query_db(backend.dbpath, sunzenithmax=sunzenithmax,
                   c3fitmin=c3fitmin, Ed400min=Ed400min, lsed400max=lsed400max)

# read timestamps, day-of-year, and days since deployment for plotting purposes
times = [datetime.datetime.strptime(r['timestamp'], '%Y-%m-%d %H:%M:%S') for r in results]
doy = [t.timetuple().tm_yday for t in times]
deployment_doy = np.min(doy)
dep_day = [d-deployment_doy for d in doy]
Ed400 = [r['Ed400'] for r in results]  # downwelling irradiance at 400 nm
LsEd400 = [r['Ls_Ed_400'] for r in results]  # ratio of sky radiance over downwelling irradiance ratio at 400nm

# keep an eye on how many observations are left for a given day, to avoid skewing results
udoy = np.unique(doy)
nobs = [len(np.argwhere(doy==d)) for d in udoy]

# plot results
pl.figure()
pl.subplot(2,1,1)
pl.plot(doy, Ed400, '.b', markersize=2, alpha=0.5, label='Ed(400) (brightness)')  # irradiance over time
pl.plot(udoy, nobs, ':k', markersize=2, alpha=0.5, label='Number of obs.')  # irradiance over time

pl.tick_params(axis='x', labelbottom=False)
pl.ylabel('Ed(400) [W/m2 nm]')
pl.legend(loc=1)
pl.subplot(2,1,2)
pl.plot(doy, LsEd400, '.g', markersize=2, alpha=0.5, label='Ls/Ed(400) (cloudiness)')  # cloudiness over time
pl.ylabel('Ls(400)/Ed(400) [1/sr]')
pl.xlabel('Day of year')
pl.legend(loc=1)
```

If your data selections were appropriate, we will have data spanning the whole deployment period but focussed on the highest Ed and lowest Ls/Ed values. Seasonality in the overall downwelling irradiance intensity will be visible; the sensors were first deployed in early April. You should find that the majority of observations where Ed(400) > 500 W/m2 nm show low sky radiance (measured at an azimuth angle 135¬∞ from the solar azimuth) compared to the total downwelling irradiance. 

## Inspecting trends in model fits

We can now look at the slope of 3C-modelled Ed against the observed Ed as an indicator of how well water-leaving reflectance could be fit for a given sample. Note that that the 3C model is highly adaptive and can usually produce a reasonable-looking reflectance spectrum even if the Ed sensor signal was already degraded. Thus, this indicator is only used to remove obvious outliers from the data set. We will remove points with a regression slope < 0.9 and days on which we have very few observations (e.g. < 100), as this likely indicates unfavourable observation conditions. These are likely affected by more extreme perturbations such as rain or suboptimal sensor geometry due to the movement of the ship. 

Another metric follows from comparing the ideal sky radiance model for Ed and the observed Ed, again, using the slope of linear regression between modelled and observed spectra. Like the 3C results, these have been pre-computed here. Values near 1 indicate that the observed Ed was close to the blue sky model predicted for the time and location. Unknown variables such as relative humidity, air pressure and ozone cause uncertainty in producing this comparison, so values deviating from 1 do not necessarily mean overcast conditions or sensor fouling, but rather to short-term variability in atmospheric conditions. Clear downward trends are, however, indicative of sensor degradation or fouling. 


```python
# select data with appropriate thresholds for sunny conditions
sunzenithmax = 50.0  # maximum allowed sun zenith angle (quality control, ensure sufficient light intensity)
c3fitmin = 0.9       # minimum selection threshold for fitting of Rrs to observation (quality control)
Ed400min = 500       # minimum downwelling irradiance intensity (ensure sufficient light intensity)
lsed400max = 0.006   # maximum value of the Lsky/Ed ratio [1/sr] at 400 nm, to select clear-sky conditions
# retrieve results from local database
results = backend.query_db(backend.dbpath, sunzenithmax=sunzenithmax,
                   c3fitmin=c3fitmin, Ed400min=Ed400min, lsed400max=lsed400max)

# read relevant variables for plots 
times = [datetime.datetime.strptime(r['timestamp'], '%Y-%m-%d %H:%M:%S') for r in results]
doy = [t.timetuple().tm_yday for t in times]

# filter out days that contribute relatively few observations
min_nobs = 100
udoy = np.unique(doy)
selection = [d for d in udoy if len(np.argwhere(doy==d)) >= min_nobs]
times =   [t for i,t in enumerate(times) if doy[i] in selection]
results = [r for i,r in enumerate(results) if doy[i] in selection]
doy =     [d for i,d in enumerate(doy) if doy[i] in selection]

deployment_doy = np.min(doy)
dep_day = [d-deployment_doy for d in doy]

Ed3C_slope = [r['Ed3C_slope'] for r in results]
EdGC_slope = [r['EdGC_slope'] for r in results]

# sort results by observation day to make boxplots
boxplotarrays3C = []
boxplotarraysGC = []
for d in np.unique(dep_day):
    boxplotarrays3C.append(np.array(Ed3C_slope)[np.argwhere(dep_day == d)])
    boxplotarraysGC.append(np.array(EdGC_slope)[np.argwhere(dep_day == d)])

# plot first data series (slope of regression of Ed in 3C model vs observed Ed) and fit trend line
props = dict(color='c')
fig = pl.figure()
pl.subplot(2,1,1)
slope, intercept, r_value, p_value, stderr = stats.linregress(dep_day, Ed3C_slope)
#pl.plot(np.unique(dep_day), intercept + np.unique(dep_day) * slope, '-c', label='3C Ed fit slope & trend')
pl.plot(np.arange(dep_day[0], dep_day[-1]+1, 1),
        intercept + np.arange(dep_day[0], dep_day[-1]+1, 1) * slope,
        '-c', label='daily 3C Ed fit slope & overall trend ({0:1.4f}%/day)'.format(100.0*slope))
bp1 = pl.boxplot(boxplotarrays3C, showfliers=False, boxprops=props, medianprops=props,
            whiskerprops=props, capprops=props, positions=np.unique(dep_day), manage_ticks=False)
pl.ylim(0.975, 1.025)
pl.legend(loc=1)

# plot second data series (slope of regression of Ed in Greg & Carder 1990 model vs observed Ed) and fit trend line
pl.subplot(2,1,2)
props = dict(color='r')
slope, intercept, r_value, p_value, stderr = stats.linregress(dep_day, EdGC_slope)
pl.plot(np.arange(dep_day[0], dep_day[-1]+1, 1),
        intercept + np.arange(dep_day[0], dep_day[-1]+1, 1) * slope,
        '-r', label='daily GC90 Ed fit slope & overall trend ({0:1.4f}%/day)'.format(100.0*slope))
bp2 = pl.boxplot(boxplotarraysGC, showfliers=False, boxprops=props, medianprops=props,
           whiskerprops=props, capprops=props, positions=np.unique(dep_day), manage_ticks=False)
pl.ylim(0.5, 1.5)
pl.legend(loc=1)
pl.xlabel('Days from deployment')
#pl.ylabel('Linear regression slope (modelled vs observed E$_d$)')
_ = fig.text(0.03, 0.5, 'Linear regression slope (modelled vs observed E$_d$)',
        ha='center', va='center', rotation='vertical')
```

We now observe a plausible degradation of the Ed signal over time in the order of -0.05% per day, compared to modelled values (lower panel). We know the sensors were regularly cleaned so this is likely associated with sensor degradation. We can also see that the quality of the water-leaving reflectance fit (upper panel) is degrading in the second half of the deployment and during the summer period, which might be due to sensor performance, measurement conditions or limitations of the model itself to reproduce water colour, which is not so relevant here. 

The fit of the data against the GC90 model is still quite variable and rather sensitive to the initial data selection of 'clear skies'. You can explore this again by choosing different threshold values - if you select only very bright illumination conditions you will end up exclusively with observations from the middle of summer. 

To reduce some of the uncertainties in this approach, it would be very useful to know when cleaning events took place. As mentioned, fitting of the Ed data against the Gregg & Carder 1990 model could be improved by recording atmospheric parameters such as relative humidity and air temperature, which greatly influence the atmospheric transmission and can be achieved with low cost sensors.

The main lesson here is that we _can_ observe gradual signal degration over time through comparison of long deployment data to models, and we can determine an acceptable rate of decay before we take the sensors down for calibration, with the potential to save time, data and money while ensureing good data quality.
